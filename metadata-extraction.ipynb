# Input JSON preparation

This notebook takes a list of json files, one corresponding to one repository within an organization, removes from the list the repositories that contain ontologies or websites, and merge the remaining ones into one JSON file.

```python
import os
import json
```

## Clear input JSONs
Removing names of github repositories containing webpages or ontologies from the list of repositories within the oeg-upm organisation

```python
##Â quitar que tengan en nombre web, website, ontology
path = '/Users/aiglesias/GitHub/oeg-software-graph/data/somef-data/single-json/'
json_file_names = os.listdir(path)
```

```python
files_to_delete = []
for file in json_file_names:
    if "web" in file or "ontology" in file:
        files_to_delete.append(file)

clean_json = [file for file in json_file_names if file not in files_to_delete]
clean_json
```

## Merge individual json into one
To facilitate posterior translation to KG

```python
merged_json = []
for file in clean_json:
    filename = path + file
    with open(filename, 'r') as infile:
        #print(json.load(infile))
        merged_json.append(json.load(infile))
    
    with open('/Users/aiglesias/GitHub/oeg-software-graph/data/somef.json', 'w') as out_json:
        json.dump(merged_json, out_json)
```
